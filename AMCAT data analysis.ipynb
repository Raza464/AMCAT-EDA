### Observations

- The 'Unnamed: 0' column is irrelevant, so we will drop it.
- DOB and DOJ columns should be converted to datetime.
- Although DOL is also a date column but it cannot be converted to datetime, due to the presence of text values 'present', which probably means the person is still working at the company.
- CollegeTier, CollegeCityTier are categorical columns, so we will convert them to object type.
- In columns like 'Domain', 'ComputerProgramming', 'ElectronicsAndSemicon'... We have many negative values, which probably because any one of these domains are aplicable. We will keep them as it's.

```python
# Drop Unnamed collumn
df.drop(columns = 'Unnamed: 0', inplace = True)
df.head(2)
```

```python
# convert DOJ and DOB columns to Datetime type and CollegeTier, CollegeCityTier to object type.

df['DOJ'] = pd.to_datetime(df['DOJ'])
df['DOB'] = pd.to_datetime(df['DOB'])

df['CollegeTier'] = df['CollegeTier'].astype(str)
df['CollegeCityTier'] = df['CollegeCityTier'].astype(str)

df[['DOJ', 'DOB', 'CollegeTier', 'CollegeCityTier']].dtypes
```

```python
cat_data = df.select_dtypes(include = 'object')
num_data = df.select_dtypes(include = 'number')
```

```python
num_data.drop(columns = ['ID', 'CollegeID', '12graduation', 'GraduationYear'], inplace = True)

cat_data.drop(columns = 'DOL', inplace = True)
```

### Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.”
Test this claim with the data given to you.

```python
designation = ['programmer analyst', 'software engineer', 'hardware engineer', 'associate engineer']
```

```python
cse = df[(df['Specialization'] == 'computer engineering') & (df['Designation'].isin(designation))]

filtered_data = cse[cse['DOJ'].dt.year == cse['GraduationYear']]

sns.histplot(filtered_data['Salary'], bins = 20, kde = True, color = palette[0], edgecolor = palette[1])
plt.title('Distribution of Salaries for  Fresh CSE Graduates who took any of the specified designation', fontsize = 15)
plt.xlabel('Salary', fontsize = 12)
plt.ylabel('Frequency', fontsize = 12)

plt.show()
```

```python
filtered_data.shape
```

```python
filtered_data[filtered_data['Salary'] > 1000000]
```

```python
# There is only one candidate earning above 10lacs, which is affecting the salary distribution. 
# Let's remove this outlier for precise inference.
filtered_data = filtered_data[filtered_data['Salary'] < 1000000]
```

```python
sns.histplot(filtered_data['Salary'], bins = 20, kde = True, color = palette[0], edgecolor = palette[1])
plt.title('Distribution of Salaries for  Fresh CSE Graduates who took any of the specified designation', fontsize = 15)
plt.xlabel('Salary', fontsize = 12)
plt.ylabel('Frequency', fontsize = 12)

plt.show()
```

```python
filtered_data['Salary'].agg([min, 'mean', 'median', 'std', max])
```

```python
from scipy.stats import ttest_1samp

# Hypothesized mean salary
hypothesized_mean = 275000

# Perform one-sample t-test
t_stat, p_value = ttest_1samp(filtered_data['Salary'], hypothesized_mean)
print(f'T-statistic: {t_stat}')
print(f'P-value: {p_value}')
```

#### Salary Claim Analysis:  
Mean salary: 405714  
Median salary: 400000  
T-statistic: 7.117192466633  
P-value: 5.277800286656114e-10 

**Conclusion:** The average and median salaries for Computer Science Engineering graduates who took jobs as Programming Analysts, Software Engineers, Hardware Engineers, or Associate Engineers are significantly higher than the claimed range of ₹2.5-3 lakhs. The extremely low p-value (1.60e-08) indicates that this difference is statistically significant, meaning we can confidently reject the claim made by the Times of India article.

### Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)

```python
contingency_table = pd.crosstab(df['Gender'], df['Specialization'])

contingency_table.T
```

```python
from scipy.stats import chi2_contingency

chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)
print(f'Chi-Square Statistic: {chi2_stat}')
print(f'P-value: {p_val}')
print(f'Degrees of Freedom: {dof}')
# print('Expected Frequencies:')
# print(expected)
```

#### Relationship Between Gender and Specialization:  

Chi-Square Statistic: 104.47  
P-value: 1.2453868176976918e-06
Degrees of Freedom: 45  

**Conclusion:** The extremely low p-value indicates a statistically significant relationship between gender and specialization. Therefore, the preference for specializations does depend on the candidate's gender.
